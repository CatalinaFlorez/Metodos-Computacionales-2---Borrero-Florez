```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"]=(15,5)
```

```python
def system (t,y,omega_0,drag,extra):
    x,v=y
    return np.array([v,-(omega_0**2)*x-drag*v-extra*(x**3)])
```

```python
sol = solve_ivp(system, y0=np.array([1.,0.]),t_span=(0,100),args=[0.5,0.05,20], max_step=0.1)
```

```python
plt.plot(sol.t,sol.y[0],".-")
```

```python
E=(sol.y[0]**2)*(0.5**2)+sol.y[1]**2
```

```python
plt.plot(sol.t,E,".-")
```
